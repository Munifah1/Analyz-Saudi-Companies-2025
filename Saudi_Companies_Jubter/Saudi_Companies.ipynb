# Sorting data

```python
#sorting data
Coms_Data.sort_index()
```

```python
plt.figure(figsize=(9,5))
sns.distplot(Coms_Data["Established_In"])
plt.xlabel("Established In")
```

From the above we can observe the most numbers of Saudi Companies Established between 2000 - 2018

```python
#grouping Ownership Type with Number of Employees
Coms_Data.groupby('Ownership_Type')['Number_of_Employees'].value_counts().dropna()
```

```python
age_buy_count = Coms_Data.sort_index().groupby('Ownership_Type').size().reset_index(name='Number_of_Employees')
fig = px.bar(age_buy_count, x='Ownership_Type', y='Number_of_Employees', title="Number of Employees",
             labels ={'Number_of_Employees':'Number of Employees','Ownership_Type':'Ownership Type'}, height=500, width= 700)
fig.show()
```

The Number of employees in many and various companies sure private sector has a higher rate

```python
plt.figure(figsize=(20,7))
px.scatter(Coms_Data, x="Established_In",y="Number_of_Employees",animation_frame="Ownership_Type",title="Number of Employees by Established Year ",
             labels ={'Number_of_Employees':'Number of Employees','Established_In':'Established_In'}, height=500, width= 700,color="Ownership_Type")
```

Run the button to observe the government and Foreign sector was high being in the past, but right now private companies have a higher rate in these days
